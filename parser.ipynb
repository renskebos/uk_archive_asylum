```python
def index_containing_substring(the_list, substring):
    for i, s in enumerate(the_list):
        if substring in s:
              return i
    return -1
```

```python
result_list = []

for i in tqdm(list(meta_data.id)):
    str_ = ""
    ind_ = index_containing_substring(corpus, i)
    if (ind_ >= 0):
        str_ = corpus[ind_]
    result_list.append((i, str_))
```

```python
df = pd.DataFrame(result_list, columns=['id', 'text'])
```

```python
def inside_search(a_list, search_term):
    result = list()
    for a in a_list:
        try:
            result.append(a.index('REASONS1')+len('REASONS1. '))
        except Exception:
            result.append(-1)
    return result


ind_list = inside_search(list(df_final.text), "REASONS1.")
ind_list[:10]
```

```python
new_texts = []

for i, t in enumerate(list(df_final.text)):
    new_texts.append(t[ind_list[i]:])

```

```python
#import sys
#!{sys.executable} -m pip install cleantext
#from cleantext import clean

def clean_text(text):
    text_encode = text.encode(encoding="ascii", errors="ignore")
    text_decode = text_encode.decode()
    text_c = " ".join([word for word in text_decode.split()])
    text_c = text_c.replace('~~~~0~~~~', '')
    text_c = text_c.lower()
    return text_c
```

```python
clean_texts = []
for t in tqdm(new_texts):
    try:
        clean_texts.append(clean_text(t))
    except Exception:
        print('error')
        break
```
